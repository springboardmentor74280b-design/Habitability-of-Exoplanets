```python
# ===============================
# PCA & t-SNE for visualization
# ===============================
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd

# Load your dataset (replace the path with your CSV file path)
df = pd.read_csv("cleaned_data.csv")

# Check the first few rows
df.head()

# -------------------------------
# 1. Prepare features and target
# -------------------------------
X = df.drop(columns=['P_HABITABLE'])
y = df['P_HABITABLE']

# -------------------------------
# 2. Standardize features
# -------------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ===============================
# PCA (2D)
# ===============================
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plot PCA
plt.figure(figsize=(6,5))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=y, palette={0:'red',1:'green',2:'orange'})
plt.title("PCA - 2D Projection")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()

# ===============================
# t-SNE (2D)
# ===============================
tsne = TSNE(n_components=2, random_state=42, learning_rate=200, perplexity=30)
X_tsne = tsne.fit_transform(X_scaled)

# Plot t-SNE
plt.figure(figsize=(6,5))
sns.scatterplot(x=X_tsne[:,0], y=X_tsne[:,1], hue=y, palette={0:'red',1:'green',2:'orange'})
plt.title("t-SNE - 2D Projection")
plt.xlabel("t-SNE Dim 1")
plt.ylabel("t-SNE Dim 2")
plt.show()
```
